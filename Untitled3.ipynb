```python
import numpy as np
import tensorflow as tf
%pwd
def save_load_pickle(pickfile, obj=None, save=True):
  import pickle
  if save:

    with open(pickfile, 'wb') as f:
        pickle.dump(obj, f) #, protocol=pickle.HIGHEST_PROTOCOL

  else:
    with open(pickfile, 'rb') as f:
        return pickle.load(f)
```
